```python
import pandas as pd
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, TimeoutException

# --- DANH SÁCH CÁC MÃ CÔNG TY CẦN CRAWL ---
TICKER_LIST_STRING = """
SEP
THU
A32
AAA
AAH
AAM
AAT
AAV
ABC
ABR
ABS
ABT
ACC
ACE
ACG
ACL
ACM
ACS
ACV
ADC
ADG
ADP
ADS
AFX
AG1
AGF
AGG
AGM
AGP
AGX
ALT
ALV
AMC
AMD
AME
AMP
AMS
AMV
ANT
ANV
APC
APF
APH
API
APL
APP
APT
ARM
ASA
ASG
ASM
ASP
AST
ATA
ATB
ATG
ATS
AVC
AVF
AVG
B82
BAF
BAL
BAX
BBC
BBH
BBM
BBS
BBT
BCA
BCB
BCC
BCE
BCF
BCG
BCM
BCO
BCP
BCR
BCV
BDB
BDG
BDT
BDW
BED
BEL
BFC
BGE
BGW
BHA
BHC
BHG
BHK
BHN
BHP
BIG
BII
BIO
BKC
BKG
BLF
BLN
BLT
BMC
BMD
BMF
BMG
BMJ
BMN
BMP
BMV
BNA
BNW
BOT
BPC
BQB
BRC
BRR
BRS
BSA
BSC
BSD
BSG
BSH
BSL
BSP
BSQ
BSR
BST
BT1
BT6
BTB
BTD
BTG
BTH
BTN
BTP
BTS
BTT
BTU
BTV
BTW
BVG
BVL
BVN
BWA
BWE
BWS
BXH
C12
C21
C22
C32
C47
C4G
C69
C92
CAB
CAD
CAG
CAN
CAP
CAR
CAT
CBI
CBS
CC1
CC4
CCA
CCC
CCI
CCL
CCM
CCP
CCR
CCT
CCV
CDC
CDG
CDH
CDN
CDO
CDP
CDR
CE1
CEG
CEN
CEO
CET
CFM
CFV
CGV
CH5
CHC
CHP
CHS
CI5
CIA
CID
CIG
CII
CIP
CJC
CK8
CKA
CKD
CKG
CKV
CLC
CLG
CLH
CLL
CLM
CLW
CLX
CMC
CMD
CMF
CMG
CMI
CMK
CMM
CMN
CMP
CMS
CMT
CMV
CMW
CMX
CNA
CNC
CNG
CNN
CNT
COM
CPA
CPC
CPH
CPI
CQN
CQT
CRC
CRE
CSC
CSM
CST
CSV
CT3
CT6
CTA
CTB
CTC
CTD
CTF
CTI
CTN
CTP
CTR
CTT
CTW
CTX
CVN
CVT
CX8
CYC
D11
D17
D2D
DAC
DAD
DAE
DAG
DAH
DAN
DAS
DAT
DBC
DBD
DBM
DBT
DC1
DC2
DC4
DCF
DCG
DCH
DCL
DCM
DCR
DCS
DCT
DDG
DDH
DDM
DDN
DDV
DFC
DFF
DGC
DGT
DGW
DHA
DHB
DHC
DHD
DHG
DHM
DHN
DHP
DHT
DIC
DID
DIG
DIH
DKC
DL1
DLD
DLG
DLR
DLT
DM7
DMC
DMN
DMS
DNA
DNC
DND
DNE
DNH
DNL
DNM
DNN
DNP
DNT
DNW
DOC
DOP
DP1
DP2
DP3
DPC
DPG
DPH
DPM
DPP
DPR
DPS
DQC
DRC
DRG
DRH
DRI
DRL
DS3
DSD
DSG
DSN
DSP
DST
DTA
DTB
DTC
DTD
DTE
DTG
DTH
DTI
DTK
DTL
DTP
DTT
DUS
DVC
DVG
DVM
DVN
DVP
DVW
DWC
DWS
DXG
DXL
DXP
DXS
DXV
DZM
E1CEP
E29
EBS
ECI
EIC
EID
EIN
ELC
EME
EMG
EMS
EPC
EPH
EVE
EVG
FBA
FBC
FCC
FCM
FCN
FCS
FDC
FGL
FHN
FHS
FIC
FID
FIR
FIT
FLC
FMC
FOC
FOX
FPT
FRC
FRM
FRT
FSO
FT1
FTI
FTM
G20
G36
GAB
GAS
GCB
GCF
GDA
GDT
GDW
GEE
GEG
GER
GEX
GGG
GH3
GHC
GIC
GIL
GKM
GLC
GLT
GLW
GMA
GMC
GMD
GMH
GMX
GND
GPC
GSM
GSP
GTA
GTD
GTS
GTT
GVR
GVT
H11
HAD
HAF
HAG
HAH
HAI
HAM
HAN
HAP
HAR
HAS
HAT
HAV
HAX
HBC
HBD
HBH
HC1
HC3
HCB
HCC
HCD
HCI
HCT
HD2
HD6
HD8
HDA
HDC
HDG
HDM
HDO
HDP
HDS
HDW
HEC
HEJ
HEP
HES
HEV
HFB
HFC
HFX
HGM
HGT
HHC
HHG
HHN
HHP
HHS
HHV
HID
HIG
HII
HIO
HJC
HJS
HKB
HKT
HLA
HLB
HLC
HLD
HLO
HLS
HLT
HLY
HMC
HMG
HMH
HMR
HMS
HNA
HNB
HND
HNF
HNG
HNI
HNM
HNP
HNR
HOM
HOT
HPB
HPD
HPG
HPH
HPI
HPM
HPP
HPT
HPW
HPX
HQC
HRB
HRC
HRT
HSG
HSI
HSL
HSM
HSP
HSV
HT1
HTC
HTE
HTG
HTI
HTL
HTM
HTN
HTP
HTT
HTV
HU1
HU3
HU4
HU6
HUB
HUG
HUT
HVG
HVH
HVN
HVT
HVX
HWS
IBC
IBD
ICC
ICF
ICG
ICI
ICN
ICT
IDC
IDI
IDJ
IDP
IDV
IFS
IHK
IJC
ILA
ILB
ILC
ILS
IME
IMP
IN4
INC
ING
INN
IRC
ISG
ISH
IST
ITA
ITC
ITD
ITQ
ITS
JOS
JVC
KAC
KBC
KCB
KCE
KDC
KDH
KDM
KGM
KHD
KHG
KHL
KHP
KHS
KHW
KIP
KKC
KLF
KMR
KMT
KOS
KPF
KSB
KSD
KSF
KSH
KSQ
KST
KSV
KTC
KTL
KTS
KTT
KTW
KVC
L10
L12
L14
L18
L35
L40
L43
L44
L45
L61
L62
L63
LAF
LAI
LAS
LAW
LBC
LBE
LBM
LCC
LCD
LCG
LCM
LCS
LDG
LDP
LDW
LEC
LG9
LGC
LGL
LGM
LHC
LHG
LIC
LIG
LIX
LKW
LLM
LM3
LM7
LM8
LMC
LMH
LMI
LNC
LO5
LPT
LQN
LSG
LSS
LTC
LTG
LUT
M10
MA1
MAC
MAS
MBG
MBN
MCC
MCF
MCG
MCH
MCM
MCO
MCP
MDA
MDC
MDF
MDG
MEC
MED
MEF
MEL
MES
MFS
MGC
MGG
MGR
MH3
MHC
MHL
MIC
MIE
MIM
MKP
MKV
MLC
MLS
MML
MNB
MND
MPC
MPT
MPY
MQB
MQN
MRF
MSH
MSN
MSR
MST
MTA
MTB
MTC
MTG
MTH
MTL
MTP
MTS
MTV
MVB
MVC
MVN
MWG
NAC
NAF
NAG
NAP
NAS
NAU
NAV
NAW
NBB
NBC
NBE
NBP
NBT
NBW
NCG
NCS
NCT
ND2
NDC
NDF
NDN
NDP
NDT
NDW
NDX
NED
NEM
NET
NFC
NGC
NHA
NHC
NHH
NHP
NHT
NHV
NJC
NKG
NLG
NLS
NNC
NNT
NO1
NOS
NQB
NQN
NQT
NRC
NS2
NSC
NSG
NSH
NSL
NSS
NST
NT2
NTB
NTC
NTF
NTH
NTL
NTP
NTT
NTW
NUE
NVL
NVP
NVT
NWT
NXT
OCH
ODE
OIL
ONE
ONW
OPC
PAC
PAI
PAN
PAP
PAS
PAT
PBC
PBP
PBT
PC1
PCC
PCE
PCF
PCG
PCH
PCM
PCN
PCT
PDB
PDC
PDN
PDR
PDV
PEC
PEG
PEN
PEQ
PET
PFL
PGC
PGD
PGN
PGS
PGT
PGV
PHC
PHH
PHN
PHP
PHR
PIA
PIC
PID
PIS
PIT
PIV
PJC
PJS
PJT
PLA
PLC
PLE
PLO
PLP
PLX
PMB
PMC
PMG
PMJ
PMP
PMS
PMT
PMW
PNC
PND
PNG
PNJ
PNP
PNT
POB
POM
POS
POT
POV
POW
PPC
PPE
PPH
PPI
PPP
PPS
PPT
PPY
PQN
PRC
PRO
PRT
PSB
PSC
PSD
PSE
PSG
PSH
PSL
PSN
PSP
PSW
PTB
PTC
PTD
PTE
PTG
PTH
PTL
PTO
PTP
PTS
PTT
PTV
PTX
PV2
PVA
PVB
PVC
PVD
PVE
PVG
PVH
PVL
PVM
PVO
PVP
PVS
PVT
PVV
PVX
PVY
PWA
PWS
PX1
PXA
PXC
PXI
PXL
PXM
PXS
PXT
QBS
QCC
QCG
QHD
QHW
QNC
QNP
QNS
QNT
QNU
QNW
QPH
QSP
QST
QTC
QTP
RAL
RAT
RBC
RCC
RCD
RCL
RDP
REE
RIC
RTB
S12
S27
S4A
S55
S72
S74
S96
S99
SAB
SAC
SAF
SAL
SAM
SAP
SAS
SAV
SB1
SBA
SBB
SBD
SBG
SBH
SBL
SBM
SBR
SBT
SBV
SC5
SCC
SCD
SCG
SCI
SCJ
SCL
SCO
SCR
SCS
SCY
SD1
SD2
SD3
SD4
SD5
SD7
SD8
SD9
SDA
SDB
SDC
SDD
SDG
SDJ
SDK
SDN
SDP
SDT
SDU
SDV
SDX
SDY
SEA
SEB
SED
SFC
SFG
SFI
SFN
SGC
SGD
SGH
SGI
SGN
SGP
SGR
SGS
SGT
SHA
SHC
SHE
SHG
SHI
SHN
SHP
SID
SIG
SII
SIP
SIV
SJ1
SJC
SJD
SJE
SJF
SJG
SJM
SJS
SKG
SKH
SKN
SKV
SLS
SMA
SMB
SMC
SMN
SMT
SNC
SNZ
SP2
SPB
SPC
SPD
SPH
SPI
SPM
SPV
SQC
SRA
SRB
SRC
SRF
SRT
SSC
SSF
SSG
SSH
SSM
SSN
ST8
STC
STG
STH
STK
STL
STP
STS
STT
STW
SVC
SVD
SVG
SVH
SVI
SVN
SVT
SWC
SZB
SZC
SZE
SZG
SZL
TA6
TA9
TAL
TAN
TAR
TAW
TB8
TBC
TBD
TBH
TBR
TBT
TBW
TBX
TCD
TCH
TCJ
TCK
TCL
TCM
TCO
TCR
TCT
TCW
TDB
TDC
TDF
TDG
TDH
TDM
TDP
TDS
TDT
TDW
TED
TEG
TEL
TET
TFC
TGG
TGP
TH1
THB
THD
THG
THM
THN
THP
THS
THT
THW
TID
TIE
TIG
TIP
TIS
TIX
TJC
TKA
TKC
TKG
TKU
TL4
TLD
TLG
TLH
TLI
TLP
TLT
TMB
TMC
TMG
TMP
TMS
TMT
TMW
TMX
TN1
TNA
TNB
TNC
TNG
TNH
TNI
TNM
TNP
TNS
TNT
TNW
TOP
TOS
TOT
TOW
TPC
TPH
TPP
TPS
TQN
TQW
TR1
TRA
TRC
TRS
TRT
TS3
TS4
TSA
TSB
TSC
TSD
TSG
TSJ
TST
TT6
TTA
TTB
TTC
TTD
TTE
TTF
TTG
TTH
TTL
TTN
TTP
TTS
TTT
TTZ
TUG
TV1
TV2
TV3
TV4
TV6
TVA
TVD
TVG
TVH
TVM
TVN
TVT
TW3
TXM
TYA
UCT
UDC
UDJ
UDL
UEM
UIC
UMC
UNI
UPC
UPH
USC
USD
V11
V12
V15
V21
VAF
VAV
VBC
VBG
VBH
VC1
VC2
VC3
VC5
VC6
VC7
VC9
VCA
VCC
VCE
VCF
VCG
VCM
VCP
VCR
VCS
VCT
VCW
VCX
VDB
VDL
VDN
VDP
VDT
VE1
VE2
VE3
VE4
VE8
VE9
VEA
VEC
VEF
VES
VET
VFC
VFG
VFR
VGC
VGG
VGI
VGL
VGP
VGR
VGS
VGT
VGV
VHC
VHD
VHE
VHF
VHG
VHH
VHL
VHM
VIC
VID
VIE
VIF
VIH
VIM
VIN
VIP
VIR
VIT
VIW
VJC
VKC
VKP
VLA
VLB
VLC
VLF
VLG
VLP
VLW
VMA
VMC
VMD
VMG
VMK
VMS
VMT
VNA
VNB
VNC
VNE
VNF
VNG
VNH
VNI
VNL
VNM
VNP
VNS
VNT
VNX
VNY
VNZ
VOC
VOS
VPA
VPC
VPD
VPG
VPH
VPI
VPR
VPS
VPW
VQC
VRC
VRE
VRG
VSA
VSC
VSE
VSF
VSG
VSH
VSI
VSM
VSN
VST
VTA
VTB
VTC
VTD
VTE
VTG
VTH
VTI
VTJ
VTK
VTL
VTM
VTO
VTP
VTQ
VTR
VTS
VTV
VTX
VTZ
VVN
VVS
VW3
VWS
VXB
VXP
VXT
WCS
WSB
WTC
X20
X26
X77
XDH
XHC
XLV
XMC
XMD
XMP
XPH
YBC
YBM
YEG
YTC
"""
TICKERS = [ticker.strip() for ticker in TICKER_LIST_STRING.split('\n') if ticker.strip()]

# --- CÁC CHỈ TIÊU CẦN LẤY ---
TARGET_CRITERIA = {
    # Bảng Cân Đối Kế Toán (BS)
    'A- TÀI SẢN NGẮN HẠN': 'Tài sản ngắn hạn',
    'TỔNG CỘNG TÀI SẢN': 'Tổng cộng tài sản',
    'C. NỢ PHẢI TRẢ': 'Nợ phải trả',
    'I. Nợ ngắn hạn': 'Nợ ngắn hạn',
    '11. Lợi nhuận sau thuế chưa phân phối': 'Lợi nhuận sau thuế chưa phân phối',
    
    # Báo Cáo Kết Quả Kinh Doanh (IS)
    '3. Doanh thu thuần về bán hàng và cung cấp dịch vụ': 'Doanh thu thuần',
    '15. Tổng lợi nhuận kế toán trước thuế': 'Lợi nhuận trước thuế',
    '18. Lợi nhuận sau thuế thu nhập doanh nghiệp': 'Lợi nhuận sau thuế',
    'Trong đó: Chi phí lãi vay': 'Chi phí lãi vay (IS)', # Lấy từ Báo cáo KQKD
    
    # Báo Cáo Lưu Chuyển Tiền Tệ (CF)
    'I. Lưu chuyển tiền thuần từ hoạt động kinh doanh': 'Lưu chuyển tiền thuần (HDKD)',
    '- Chi phí lãi vay': 'Chi phí lãi vay (CF)' # Lấy từ Báo cáo LCTT
}

# Các cột cuối cùng trong file Excel
FINAL_COLUMNS = [
    'Firm', 'Year',
    'Tài sản ngắn hạn', 'Tổng cộng tài sản', 'Nợ phải trả', 'Nợ ngắn hạn', 
    'Lợi nhuận sau thuế chưa phân phối', 'Doanh thu thuần', 
    'Lợi nhuận trước thuế', 'Lợi nhuận sau thuế', 
    'Chi phí lãi vay (IS)', 'Chi phí lãi vay (CF)',
    'Lưu chuyển tiền thuần (HDKD)'
]

# --- CÁC HÀM HỖ TRỢ ---

def get_elements_value(items, convert_numeric=True):
    """Lấy text từ các element và chuyển đổi sang số."""
    _ls = []
    if convert_numeric == False:
        for x in items:
            _ls.append(x.text)
    else:
        for x in items:
            try:
                num_str = x.text.strip().replace(',', '')
                if num_str == '' or num_str == '-':
                    _ls.append(None) # Xử lý chuỗi rỗng hoặc gạch nối
                else:
                    _ls.append(float(num_str)) # Dùng float cho dữ liệu tài chính
            except ValueError:
                _ls.append(x.text) # Giữ nguyên text nếu không thể chuyển đổi
    return _ls

def get_report_url(ticker, year, report_type):
    """Tạo URL cho report."""
    if report_type.upper() == 'BS':
        return f"https://cafef.vn/du-lieu/bao-cao-tai-chinh/{ticker}/bsheet/{year}/0/0/0/0/ket-qua-hoat-dong-kinh-doanh-.chn"
    if report_type.upper() == 'IS':
        return f"https://cafef.vn/du-lieu/bao-cao-tai-chinh/{ticker}/incsta/{year}/0/0/0/0/0/ket-qua-hoat-dong-kinh-doanh-.chn"
    if report_type.upper() == 'CF':
        return f"https://cafef.vn/du-lieu/bao-cao-tai-chinh/{ticker}/cashflow/{year}/0/0/0/0/ket-qua-hoat-dong-kinh-doanh-.chn"
    return ""

# --- HÀM CRAWL CHÍNH ---

def scrape_data_for_ticker(driver, ticker, all_data_records):
    """
    Scrape tất cả dữ liệu (BS, IS, CF) cho MỘT ticker 
    và thêm vào list `all_data_records`.
    """
    print(f"--- Bắt đầu scrape {ticker} ---")
    
    ticker_data_by_year = {}
    
    for report_type in ['BS', 'IS', 'CF']:
        print(f"  - Report {report_type}")
        year = 2025 # Bắt đầu từ 2025 để lấy trang có dữ liệu mới nhất (bao gồm 2024)
        runable = True
        seen_years = set() 
        
        while(runable):
            url = get_report_url(ticker, year, report_type)
            
            print(f"    - Đang tải URL: {url}")
            try:
                driver.get(url)
            except TimeoutException:
                print(f"    - LỖI: Trang tải quá 15 giây. Bỏ qua {ticker} {report_type} năm {year}.")
                runable = False 
                continue 

            time.sleep(1.5) # Chờ trang tải
            
            try:
                driver.find_element(By.ID, "tableContent")
            except NoSuchElementException:
                print(f"    - Không tìm thấy dữ liệu cho {ticker} {report_type} năm {year}. Dừng report này.")
                runable = False
                continue
            
            # 1. Lấy danh sách tên chỉ tiêu (các hàng)
            name_elements = driver.find_elements(By.XPATH, "//*[@id='tableContent']/tbody/tr/td[1]")
            criteria_texts = [el.text.strip() for el in name_elements]
            
            # 2. Lấy danh sách các năm có sẵn (các cột)
            year_elements = driver.find_elements(By.XPATH, "//*[@id='tblGridData']/tbody/tr/td")
            index_cols = {}
            i = 1
            for item in year_elements:
                str_item = item.text.strip()
                if str_item.isnumeric():
                    index_cols[str_item] = i # Key=năm, Value=index cột
                i += 1
            
            if not index_cols:
                print(f"    - Không tìm thấy cột năm nào cho {ticker} {report_type} tại URL {year}. Dừng report này.")
                runable = False
                continue

            index_cols_sorted = dict(sorted(index_cols.items(), reverse=True))
            oldest_year_on_page = 0
            new_years_found_on_this_page = False # Đổi tên biến để rõ ràng hơn

            for year_str, col_idx in index_cols_sorted.items():
                try:
                    current_year_int = int(year_str)
                except ValueError:
                    print(f"      - Bỏ qua cột không phải năm: {year_str}")
                    continue
                
                # === THAY ĐỔI MỚI (1): Chỉ lấy đến năm 2024 ===
                if current_year_int > 2024:
                    print(f"      - Bỏ qua năm {year_str} (lớn hơn 2024).")
                    continue
                
                # Gán năm cũ nhất trên trang (sẽ dùng để lùi trang)
                # Đảm bảo nó là năm <= 2024
                if oldest_year_on_page == 0:
                    oldest_year_on_page = current_year_int
                else:
                    oldest_year_on_page = min(oldest_year_on_page, current_year_int)

                if year_str in seen_years:
                    continue 
                
                seen_years.add(year_str) 
                
                year_data_elements = driver.find_elements(By.XPATH, f"//*[@id='tableContent']/tbody/tr/td[{col_idx}]")
                year_data_values = get_elements_value(year_data_elements)
                
                # === THAY ĐỔI MỚI (2): Kiểm tra xem cột này có dữ liệu hay không ===
                all_empty = True
                for val in year_data_values:
                    # Nếu có bất kỳ giá trị nào không rỗng, đánh dấu là không empty
                    if val is not None and str(val).strip() != '':
                        all_empty = False
                        break
                
                if all_empty:
                    print(f"      - Năm {year_str} không có dữ liệu, bỏ qua.")
                    continue # Bỏ qua năm này, không xử lý

                # === NẾU CHẠY ĐẾN ĐÂY, NĂM NÀY HỢP LỆ (<= 2024 và CÓ DATA) ===
                new_years_found_on_this_page = True # Đánh dấu là trang này có dữ liệu mới
                
                if year_str not in ticker_data_by_year:
                    ticker_data_by_year[year_str] = {'Firm': ticker, 'Year': year_str}
                
                # 5. Map dữ liệu vào dictionary
                for i, criteria_text in enumerate(criteria_texts):
                    if i >= len(year_data_values): continue 
                    if not criteria_text: continue 
                    
                    for target_key, out_col_name in TARGET_CRITERIA.items():
                        if criteria_text.startswith(target_key):
                            ticker_data_by_year[year_str][out_col_name] = year_data_values[i]
                            break 
            
            if not new_years_found_on_this_page: # Sử dụng biến đã đổi tên
                print(f"    - Không tìm thấy dữ liệu mới/hợp lệ trên trang này. Dừng report {report_type}.")
                runable = False
            else:
                if oldest_year_on_page == 0:
                     # Trường hợp trang này chỉ có dữ liệu > 2024 (ví dụ 2025)
                     print(f"    - Chỉ có dữ liệu > 2024. Dừng report {report_type}.")
                     runable = False
                else:
                    # Lùi về trang trước đó
                    year = oldest_year_on_page - 1
    
    all_data_records.extend(ticker_data_by_year.values())
    print(f"--- Hoàn thành {ticker}, tìm thấy {len(ticker_data_by_year)} năm hợp lệ. ---")

# --- HÀM THỰC THI CHÍNH ---
def main():
    print("Khởi tạo Chrome driver...")
    service = ChromeService(ChromeDriverManager().install())
    chrome_options = ChromeOptions()
    
    # === THAY ĐỔI MỚI: Tắt headless để test ===
    # NẾU BỊ KẸT, HÃY ĐỂ DÒNG NÀY CÓ DẤU # (chạy có giao diện) ĐỂ XEM LỖI.
    # Khi chạy ổn rồi, bạn có thể bỏ comment (xóa dấu #) dòng dưới để chạy ẩn.
    # chrome_options.add_argument("--headless=new") 
    
    chrome_options.add_argument("--log-level=3") # Giảm log rác
    driver_auto = webdriver.Chrome(service=service, options=chrome_options)
    
    # === THAY ĐỔI MỚI: Thêm timeout (15 giây) ===
    # Nếu trang không tải trong 15s, nó sẽ báo lỗi (TimeoutException)
    driver_auto.set_page_load_timeout(15)
    
    all_data_list = [] # List tổng chứa dữ liệu của TẤT CẢ các công ty
    
    for ticker in TICKERS:
        try:
            scrape_data_for_ticker(driver_auto, ticker, all_data_list)
        except Exception as e:
            print(f"!!! LỖI NGHIÊM TRỌNG khi scrape {ticker}: {e}")
            continue 
    
    driver_auto.quit()
    
    print(f"\nTổng cộng scrape được {len(all_data_list)} bản ghi Firm-Year.")
    
    if not all_data_list:
        print("Không có dữ liệu nào được scrape. Thoát.")
        return
        
    df = pd.DataFrame(all_data_list)
    
    df_final = df.reindex(columns=FINAL_COLUMNS)
    
    try:
        df_final['Year'] = pd.to_numeric(df_final['Year'])
        df_final = df_final.sort_values(by=['Firm', 'Year'], ascending=[True, False])
    except Exception as e:
        print(f"Không thể sắp xếp dữ liệu: {e}")

    output_file = "all_firm_data.xlsx"
    try:
        writer = pd.ExcelWriter(output_file, engine='xlsxwriter')
        df_final.to_excel(writer, sheet_name='FinancialData', index=False)
        writer.close() 
        print(f"ĐÃ LƯU DỮ LIỆU vào file: {output_file}")
        os.startfile(output_file) 
    except Exception as e:
        print(f"Không thể lưu file Excel (có thể do file đang mở): {e}")
        print("Đang thử lưu vào file CSV...")
        output_file_csv = "all_firm_data.csv"
        try:
            df_final.to_csv(output_file_csv, index=False, encoding='utf-8-sig')
            print(f"Đã lưu thành công vào: {output_file_csv}")
            os.startfile(output_file_csv)
        except Exception as e2:
            print(f"Lưu file CSV cũng thất bại: {e2}")

if __name__ == "__main__":
    main()

```
